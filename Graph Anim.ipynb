```python
import plotly.graph_objects as go
from loclust import parse
from scripts.utilities import trajectories_to_dataframe as traj_to_df
import pandas as pd
import numpy as np

# loads the dataset into a dataframe and generates list of x time values
dataset = pd.read_csv('./data/df_lorj.csv')
x_vals = np.sort(dataset['X'].unique(), kind='heapsort')
x_vals_str = map(lambda to_str: str(to_str), x_vals)

#continents, used for iteration
continents = []
for continent in dataset['name']:
    if continent not in continents:
        continents.append(continent)

# make figure dict structure
fig_dict = {
    "data": [],
    "layout": {},
    "frames": []
}

# fill in most of layout specific details
fig_dict["layout"]["xaxis"] = {'range':[0,300], "title": "Adjusted Fertility"}
fig_dict["layout"]["yaxis"] = {"title": "Adjusted Income", "type": "log",'range':[0,6]}
fig_dict["layout"]["hovermode"] = "closest"
fig_dict["layout"]["sliders"] = {
    "args": [
        "transition", {
            "duration": 800,
            "easing": "cubic-in-out"
        }
    ],
    "initialValue": str(x_vals[0]),
    "plotlycommand": "animate",
    "values": x_vals_str,
    "visible": True
}
# Defines theme for chart
fig_dict["layout"]["template"] = "plotly_dark"
#Update behavior 
fig_dict["layout"]["updatemenus"] = [
    {
        "buttons": [
            {
                "args": [None, {"frame": {"duration": 500, "redraw": False},
                                "fromcurrent": True, "transition": {"duration": 800,
                                                                    "easing": "quadratic-in-out"}}],
                "label": "Play",
                "method": "animate"
            },
            {
                "args": [[None], {"frame": {"duration": 0, "redraw": False},
                                  "mode": "immediate",
                                  "transition": {"duration": 0}}],
                "label": "Pause",
                "method": "animate"
            }
        ],
        "direction": "left",
        "pad": {"r": 10, "t": 87},
        "showactive": False,
        "type": "buttons",
        "x": 0.1,
        "xanchor": "right",
        "y": 0,
        "yanchor": "top"
    }
]
#Defines properties around slider behavior
sliders_dict = {
    "active": 0,
    "yanchor": "top",
    "xanchor": "left",
    "currentvalue": {
        "font": {"size": 20},
        "prefix": "Year:",
        "visible": True,
        "xanchor": "right"
    },
    "transition": {"duration": 300, "easing": "cubic-in-out"},
    "pad": {"b": 10, "t": 50},
    "len": 0.9,
    "x": 0.1,
    "y": 0,
    "steps": []
}

#make data for initial values
x_val = x_vals[0]
for continent in continents:
    dataset_by_year = dataset[dataset["X"] == x_val]
    dataset_by_year_and_cont = dataset_by_year[
        dataset_by_year["name"] == continent]

    data_dict = {
        "x": list(dataset_by_year_and_cont['ad_fert_data']),
        "y": list(dataset_by_year_and_cont["adjusted_income_data"]),
        "mode": "markers",
        "text": list(dataset_by_year_and_cont["country_name"]),
        "marker": {
            "sizemode": "area",
            "sizeref": 0.1,
            "size": list(map(lambda increm: int((50 * (increm/100)) + 15), 
                                 list(dataset_by_year_and_cont["contraceptive_data"].fillna(0)))),
            "opacity": 0
        },
        "name": continent
    }
    fig_dict["data"].append(data_dict)

# make frames
# iterates over all time values
for x_valu in x_vals_str:
    frame = {"data": [], "name": str(x_valu)}
#     for each time value add all the distinct entries
    for continent in continents:
        dataset_by_year = dataset[dataset["X"] == int(float(x_valu))]
        dataset_by_year_and_cont = dataset_by_year[
            dataset_by_year["name"] == continent]

        data_dict = {
            "x": list(dataset_by_year_and_cont["ad_fert_data"]),
            "y": list(dataset_by_year_and_cont["adjusted_income_data"]),
            "mode": "markers",
            "text": list(dataset_by_year_and_cont["country_name"]),
            "marker": {
                "sizemode": "area",
                "sizeref": 0.1,
                #Adds the size for each value, filling nan values and casting all values to int
                "size": list(map(lambda increm: int((50 * (increm/100)) + 15), 
                                 list(dataset_by_year_and_cont["contraceptive_data"].fillna(0)))),
                "opacity":0.7
            },
            "name": continent
        }
        frame["data"].append(data_dict)

    fig_dict["frames"].append(frame)
    #Make sure to match the keys in the slider step to each frame and the slider in both value and type
    slider_step = {"args": [
        [x_valu],
        {"frame": {"duration": 300, "redraw": False},
         "mode": "immediate",
         "transition": {"duration": 300}}
    ],
        "label": x_valu,
        "method": "animate"}
    sliders_dict["steps"].append(slider_step)


fig_dict["layout"]["sliders"] = [sliders_dict]

#Defines figure object from dictionary
fig = go.Figure(fig_dict)
fig.add_trace(
    go.Scatter(
        x=[1, 274],
        y=[1000000, 5],
        mode="lines",
        line=go.scatter.Line(color="red"),
        showlegend=False)
)
fig.show()
```
