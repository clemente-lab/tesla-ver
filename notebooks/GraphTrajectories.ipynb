```python
#Data Manipulation and Trajectory Parsing Libraries
import plotly.plotly as py
import pandas as pd
import numpy as np
from loclust import parse
from teslaver import utilities as utls

#plotly graphing libraries
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
from plotly.graph_objs import *
from plotly.grid_objs import Grid, Column
from plotly.tools import FigureFactory as FF

def plot_graph(path_to_trajectories, xaxis_title, yaxis_title, autorangeTF = True, slide_duration=400):
    #Defines offline plotly graphing as the designated mode
    init_notebook_mode()

    #Created Dataframe from loclust trajectories
    try:
        parsed_dataframe = utls.trajectories_to_dataframe(parse.read_trajectories(path_to_trajectories))
        xvals = [traj.time for traj in parse.read_trajectories(path_to_trajectories)]
    #TODO: add specific error handling here
    except:
        print("Invalid Path or Invalid loclust installation")
    figure = {
        'data': [],
        'layout': {},
        'frames': [],   
    }
    #define axis layout and hovering layout
    figure['layout']['xaxis'] = {'title': xaxis_title, 'autorange':autorangeTF}
    figure['layout']['yaxis'] = {'title': yaxis_title, 'autorange':autorangeTF}
    figure['layout']['hovermode'] = 'closest'
    figure['layout']['showlegend'] = True
    
    #defined layout for slider behaviors
    figure['layout']['sliders'] = {
        'args':[
            'slider.value', {
                'duration': 400,
                'ease': 'cubic-in-out'
            }
        ],
        'initialValue': min(xvals),
        'plotlycommand': xvals,
        'visible':True
    }
    #described figure buttons and menus
    figure['layout']['updatemenu'] = [
        {
            'buttons':[
                {
                    'args':[None, {'frame':{'duration':slide_duration, 'redraw':False}}],
                    'label': 'Pause',
                    'method': 'animate'
                },
                {
                    'args': [[None], {'frame': {'duration': 0, 'redraw': False}, 'mode': 'immediate',
                    'transition': {'duration': 0}}],
                    'label': 'Pause',
                    'method': 'animate'
                }
            ],
            
        }
    ]

```
